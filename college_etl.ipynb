```python
#Import Dependencies
import pandas as pd
from sqlalchemy import create_engine
from sqlalchemy.types import String
```

### Extract CSVs into DataFrames

```python
#Loading CSV files
colleges = "Resources/Colleges_and_Universities.csv"
merged_file = "Resources/MERGED2017_18_PP.csv"
```

```python
#CSV to DataFrame
colleges_df = pd.read_csv(colleges)
colleges_df.head()
```

```python
#CSV to DataFrame
merged_data = pd.read_csv(merged_file, low_memory=False)
merged_data.head()
```

```python
#Pulling the names of columns to select for a new data frame
colleges_df.columns.tolist()
```

```python
# Creating a filtered dataframe from specific columns
colleges_cols = ["IPEDSID", "NAME", "ADDRESS","CITY","STATE","ZIP","TELEPHONE","WEBSITE","TYPE","STATUS","POPULATION","LATITUDE","LONGITUDE"]
colleges_transformed= colleges_df[colleges_cols].copy()
```

```python
colleges_transformed.head()
```

```python
#Renaming comumns headings to load into pgAdmin
student_choice_new  = colleges_transformed.rename(columns={"IPEDSID": "ipedsid",
                                                          "NAME": "name",
                                                          "ADDRESS": "address",
                                                          "CITY": "city",
                                                          "STATE": "state",
                                                          "ZIP": "zip",
                                                          "TELEPHONE": "telephone",
                                                          "WEBSITE": "website",
                                                          "TYPE": "type",
                                                          "STATUS": "status",
                                                          "POPULATION": "population",
                                                          "LATITUDE": "latitude",
                                                          "LONGITUDE": "longitude"
                                                })
student_choice_new.head()
```

```python
#Cleaning dataframe: dropping dublicates in index/future primery key column
student_choice_new.drop_duplicates("ipedsid", inplace=True)
#Resetting index to ipedsid
student_choice_new.set_index("ipedsid", inplace=True)
#Dropping N/A
student_choice=student_choice_new.dropna()
student_choice.head()
```

### Transform Second DataFrame

```python
## Creating a filtered dataframe from specific columns
merged_admin_col =["UNITID","ADM_RATE","ACTCMMID","SAT_AVG","TUITIONFEE_IN",'TUITIONFEE_OUT']
#Making a copy of the new data set
merged_admin_1 = merged_data[merged_admin_col].copy()
#Renaiming heading of the column wich will be set to index
merged_admin = merged_admin_1.rename(columns={ "UNITID": "ipedsid"})
#Cleaning dataframe: dropping dublicates in index/future primery key column
merged_admin.drop_duplicates("ipedsid", inplace=True)
merged_admin.set_index("ipedsid", inplace=True)
merged_admin_clean=merged_admin.dropna()
merged_admin_clean.head()
```

```python
#Renaming comumns headings to load into pgAdmin
college_choice = merged_admin_clean.rename(columns={ "ADM_RATE": "adm_rate",
                                                          "ACTCMMID": "actcmmid",
                                                          "SAT_AVG": "sat_avg",
                                                          "TUITIONFEE_IN": "tuitionfee_in",
                                                          "TUITIONFEE_OUT": "tuitionfee_out"
                                                    })
college_choice.head()
```

### Create database connection

```python
connection_string = "postgres:postgres@localhost:5432/college_db"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
# Confirm tables 
engine.table_names()
```

### Load DataFrames into database

```python
#Loading first data frame into pgAdmin database
student_choice.to_sql(name='student_choice', con=engine, if_exists='append', index=True)
```

```python
#Loading second data frame into pgAdmin database
college_choice.to_sql(name='college_choice', con=engine, if_exists='append', index=True)
```

```python
#Confirming data has been added by querying the student_choice table
pd.read_sql_query('select * from student_choice', con=engine).head()
```
